```python
from scipy.stats import skewnorm
import matplotlib.pyplot as plt
import seaborn as sns

# Parameters
num_values = 1000
max_value = 100
skewness = 20  # Positive values for right skewed, negative values for left skewed

# Generate a skewed dataset
random_data = skewnorm.rvs(a=skewness, loc=0, scale=max_value, size=num_values)

# Normalize the dataset to the range [0, max_value]
random_data = random_data - min(random_data)  # Shift to start at 0
random_data = random_data / max(random_data)  # Normalize to range [0, 1]
random_data = random_data * max_value  # Scale to the desired range

# Create the subplots
fig, axs = plt.subplots(2, 2, figsize=(15, 12))

# Plot the original data
sns.histplot(random_data, bins=50, kde=True, color='c', edgecolor='black', alpha=0.7, ax=axs[0, 0])
axs[0, 0].set_title('Histogram of Skewed Dataset with KDE')
axs[0, 0].set_xlabel('Value')
axs[0, 0].set_ylabel('Frequency')

# Plot the transformed data (x**0.5)
sns.histplot(random_data**0.5, bins=50, kde=True, color='b', edgecolor='black', alpha=0.7, ax=axs[0, 1])
axs[0, 1].set_title('Histogram of Transformed Dataset (x**0.5) with KDE')
axs[0, 1].set_xlabel('Value')
axs[0, 1].set_ylabel('Frequency')

# Plot the transformed data (x**0.25)
sns.histplot(random_data**0.25, bins=50, kde=True, color='g', edgecolor='black', alpha=0.7, ax=axs[1, 0])
axs[1, 0].set_title('Histogram of Transformed Dataset (x**0.25) with KDE')
axs[1, 0].set_xlabel('Value')
axs[1, 0].set_ylabel('Frequency')

# Plot the transformed data (x**0.4)
sns.histplot(random_data**0.4, bins=50, kde=True, color='m', edgecolor='black', alpha=0.7, ax=axs[1, 1])
axs[1, 1].set_title('Histogram of Transformed Dataset (x**0.4) with KDE')
axs[1, 1].set_xlabel('Value')
axs[1, 1].set_ylabel('Frequency')

# Adjust layout
plt.tight_layout()
plt.show()
```

```python
# Import required modules
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('./deeplearning.mplstyle')
```

```python
# x_train is the input variable - feature (in 1000 square feet)
# y_train is the target (Price in 1000s of dollars)
x_train = np.array([1.0, 2.0])
y_train = np.array([300.0, 500.0])
print(f"x_train = {x_train}")
print(f"y_train = {y_train}")
```

```python
# Set m the number of training examples
print(f"x_train.shape = {x_train.shape}")
m = x_train.shape[0]
print(f"The number of training examples is: {m}")
```

```python
# Get the i^ith training example
i = 0
try:
    x_i = x_train[i]
    y_i = y_train[i]
    print(f"(x^({i}), y^{i}) = ({x_i}, {y_i})")
except IndexError:
    print(f"The index {i} does not exist")
```

```python
# Plot the two data points
plt.scatter(x_train, y_train, marker='x', c="r")
# Set the title
plt.title("Housing Prices")
# Y-axis label
plt.ylabel("Price (in 1000s $)")
# X-axis label
plt.xlabel("Size (1000 sqft)")
plt.show()
```

```python
w = 200
b =100
```

```python
# Compute for data points
def compute_model_output(x, w, b):
    """Computes the prediction f of a linear model

    Keyword arguments:
    x (ndarray (m,)): Data, m examples
    w,b (scalar)    : Model parameters
    Return:
    f_wb (ndarray (m,)) : model prediction
    """
    m = x.shape[0]
    f_wb = np.zeros(m)
    for i in range(m):
        f_wb[i] = w * x[i] + b

    return f_wb
```

```python
# Call and apply model
tmp_f_wb = compute_model_output(x_train, w, b,)

# Plot the prediction
plt.plot(x_train, tmp_f_wb, c="b", label="Our Prediction")
#plot the data points
plt.scatter(x_train, y_train, marker="x", c="r", label="Actual Values")
plt.title("Housing Prices")
plt.ylabel("Price (in 1000s of dollars)")
plt.xlabel("Size (1000 sqft)")
plt.legend()
plt.show()
```

```python
# Now apply to predict
w = 200
b =100
x_i = 1.2
cost_1200sqft = w * x_i + b
print(f"The cost of a house with 1200 sqft is predicted to be: {cost_1200sqft} thousand dollars")
```
